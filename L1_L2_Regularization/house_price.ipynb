```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
dataset = pd.read_csv("melb_data.csv")
dataset.head()
```

```python
dataset.shape
```

```python
data_to_use = ['Suburb', 'Rooms','Type','Method','SellerG','Regionname','Propertycount','Distance','CouncilArea',
               'Bedroom2','Bathroom','Car','Landsize','BuildingArea','Price']
dataset = dataset[data_to_use]
dataset.head()
```

```python
dataset.isna().sum()
```

```python

dataset['BuildingArea'] = dataset['BuildingArea'].fillna(dataset['BuildingArea'].mean())
dataset['Car'] = dataset['Car'].fillna(dataset['Car'].mean())
dataset['CouncilArea'] = dataset['CouncilArea'].fillna(0)
dataset.isna().sum()
```

```python
dataset = pd.get_dummies(dataset, drop_first= True).astype(int)
dataset.head()
```

```python
X = dataset.drop('Price', axis = 'columns')
y = dataset.Price
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test, y_train , y_test = train_test_split(X,y,test_size=0.3,random_state= 50) 
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
model.score(X_train,y_train)
```

```python
from sklearn import linear_model

lassso_reg = linear_model.Lasso(alpha= 20 , max_iter= 100 , tol = 0.1)

lassso_reg.fit(X_train,y_train)
```

```python
lassso_reg.score(X_test,y_test)
```

```python
lassso_reg.score(X_train,y_train)
```

```python
from sklearn.linear_model import Ridge
ridge_reg = Ridge(alpha= 20,max_iter=100,tol=0.1)
ridge_reg.fit(X_train,y_train)
```

```python
ridge_reg.score(X_test,y_test)
```

```python
ridge_reg.score(X_train,y_train)
```
