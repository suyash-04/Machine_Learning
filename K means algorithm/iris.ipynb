```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
from matplotlib import pyplot as plt
%matplotlib inline
from sklearn.datasets import load_iris

datas= load_iris()
```

```python
dir(datas)
```

```python
df = pd.DataFrame(datas.data, columns=datas.feature_names)
df.head()
```

```python
inputs = df.drop(["sepal length (cm)","sepal width (cm)"], axis="columns")
```

```python
plt.scatter(inputs["petal length (cm)"],inputs["petal width (cm)"])
```

```python
sse=[]
k_rng = range(1,10)
for k in k_rng:
    km = KMeans(n_clusters= k)
    km.fit(inputs[["petal length (cm)","petal width (cm)"]])
    sse.append(km.inertia_)
```

```python
plt.plot(k_rng,sse)
```

```python
km = KMeans(n_clusters=3)
yp= km.fit_predict(df)
```

```python
inputs['cluster'] = yp
inputs
```

```python
df0= inputs[inputs.cluster == 0]
df1= inputs[inputs.cluster == 1]
df2= inputs[inputs.cluster == 2]
```

```python
plt.scatter(df0['petal width (cm)'],df0['petal length (cm)'],color ='blue')
plt.scatter(df1['petal width (cm)'],df1['petal length (cm)'],color ='red')
plt.scatter(df2['petal width (cm)'],df2['petal length (cm)'],color ='green')
```
